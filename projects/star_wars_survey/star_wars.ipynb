# Exploring star wars survey responses

Some important columns:
* RespondentID - An anonymized ID for the respondent (person taking the survey)
* Gender - The respondent's gender
* Age - The respondent's age
* Household Income - The respondent's income
* Education - The respondent's education level
* Location (Census Region) - The respondent's location
* Have you seen any of the 6 films in the Star Wars franchise? - Has a Yes or No response
* Do you consider yourself to be a fan of the Star Wars film franchise? - Has a Yes or No response

```python
import pandas as pd

pd.set_option('display.notebook_repr_html', True)
pd.set_option('display.html.table_schema', True)
star_wars = pd.read_csv("star_wars.csv",encoding="ISO-8859-1")

star_wars = star_wars[star_wars["RespondentID"].notnull()]

star_wars.head(10)
```

```python
yes_no = {"Yes": True, "No": False}
cols = ["Have you seen any of the 6 films in the Star Wars franchise?","Do you consider yourself to be a fan of the Star Wars film franchise?"]
for each in cols:
    print(star_wars[each])
    star_wars[each] = star_wars[each].map(yes_no)
star_wars.head()
```

```python
import numpy as np

seen_dict = {
        "Star Wars: Episode I  The Phantom Menace":True,
    "Star Wars: Episode II  Attack of the Clones":True,
    "Star Wars: Episode III  Revenge of the Sith":True,
    "Star Wars: Episode IV  A New Hope":True,
    "Star Wars: Episode V The Empire Strikes Back":True,
    "Star Wars: Episode VI Return of the Jedi":True,
    np.nan: False
}

col_names = star_wars.columns[3:9].values
col_dict = {}

import numpy as np

for i, each in enumerate(col_names):
    star_wars[each] = star_wars[each].map(seen_dict)
    col_dict[col_names[i]] = "seen_{}".format(i+1)
    
star_wars = star_wars.rename(columns=col_dict)
star_wars.head()
```

```python
# clean ranking columns
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

col_dict = {}
for i, each in enumerate(star_wars.columns.values[9:15]):
    col_dict[each] = "ranking_{0}".format(i+1)
    
star_wars = star_wars.rename(columns=col_dict)
star_wars[star_wars.columns[9:15]].head()
```

```python
mean = star_wars[star_wars.columns[9:15]].mean()
print(mean)
xlabels = ["Episode 1", "Episode 2", "Episode 3", "Episode 4", "Episode 5", "Episode 6"]
%matplotlib inline
ax = mean.plot.bar()
ax.set_xticklabels(xlabels)
ax.set_title("Mean ranking of each Star Wars movie")
ax.set_ylim(0, 6)
ax.set_ylabel("Ranking (1 = best, 6 = worst)")
```

### Original movies are more highly rated than old episodes.  Episode 5 is the highest rated (because it's clearly the best)

```python
# Doing above with matplotlib directly
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
bar_positions = np.arange(1,7)
colors = ["blue", "orange", "green", "red", "purple", "brown"]
ax.bar(bar_positions,mean, width = 0.5, color=colors, tick_label=xlabels)
ax.set_title("Mean ranking of each Star Wars movie")
ax.set_ylim(0, 6)
ax.set_ylabel("Ranking (1 = best, 6 = worst)")
```

```python
# Doing above with seaborn
import seaborn as sns

with sns.axes_style("darkgrid"):
    sns.barplot(x=xlabels,y=mean)
```

```python
seen = star_wars[star_wars.columns[3:9]].sum()
ax = seen.plot.bar()
ax.set_title("Number of respondents who have seen each movie")
ax.set_xticklabels(xlabels)
```

Episodes 5 and 6 are the most seen movies, which makes sense given their high rankings

# Males vs Females

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

male_mean = males[males.columns[9:15]].mean()
female_mean = females[females.columns[9:15]].mean()


ax = male_mean.plot.bar()
ax.set_title("Ranking amongst males")
plt.show()
ax = female_mean.plot.bar()
ax.set_title("Ranking amongst females")
plt.show()
```

```python
# Combining into one plot

xlabels.insert(0,0)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(10,5))

axes[0].bar(np.arange(1,7)-0.125,male_mean,width=0.25,label="male")
axes[0].bar(np.arange(1,7)+0.125,female_mean,width=0.25,label="female")
axes[0].legend(loc="upper right")
axes[0].set_ylim(0,6)
axes[0].set_xticklabels(xlabels, rotation = 90,fontsize=8)
axes[0].set_title ("Film ratings by gender")

axes[1].bar(np.arange(1,7)-0.125,males[males.columns[3:9]].sum(),width=0.25,label="male")
axes[1].bar(np.arange(1,7)+0.125,females[females.columns[3:9]].sum(),width=0.25,label="female")
axes[1].legend(loc="upper right")
axes[1].set_xticklabels(xlabels)
plt.xticks(rotation=90)
axes[1].set_title ("Film viewings by gender")
plt.show()
    
```

```python
education_df = star_wars.groupby("Education").agg(np.mean)
education_df.reset_index(inplace=True)
education_df.head()


education_df = education_df.rename(columns={"Have you seen any of the 6 films in the Star Wars franchise?":"seen?"})
education_df.sort_values(by="seen?",ascending=True,inplace=True)

ax = education_df["seen?"].plot.bar(title="Fraction who have seen a Star Wars film")
ax.set_xticklabels(education_df["Education"].tolist())
```

### Likelihood of having seen a Star Wars movie is positively correlated with education

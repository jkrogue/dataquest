```python
# Improving plot aesthetics

import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

women_degrees = pd.read_csv("data/percent-bachelors-degrees-women-usa.csv")

women_degrees.dropna(inplace=True)

print(women_degrees[:10]["Year"])

fig, ax = plt.subplots()
ax.plot(women_degrees["Year"],women_degrees["Biology"])
ax.set_title("Percent of biology degrees given to women over the years")
ax.set_ylabel("Percent women")
ax.set_xlabel("Year")
ax.set_xticklabels(women_degrees["Year"],rotation=90)
plt.show()
```

```python
male_biology = 100 - women_degrees["Biology"]
```

```python
fig, ax = plt.subplots()
ax.plot(women_degrees["Year"],women_degrees["Biology"], label="Women", color="green")
ax.plot(women_degrees["Year"],male_biology, label="Men", color="blue")
ax.set_title("Percentage of Biology Degrees Awarded By Gender")

fig.legend(loc="upper right")
```

```python
# Hiding tick marks

fig, ax = plt.subplots()
ax.plot(women_degrees["Year"],women_degrees["Biology"], label="Women", color="green")
ax.plot(women_degrees["Year"],100-women_degrees["Biology"], label="Men", color="blue")
ax.set_title("Percentage of Biology Degrees Awarded By Gender")
ax.tick_params(left="off",bottom="off",top="off",right="off")
fig.legend(loc="upper right")
```

```python
# Hiding spines

fig, ax = plt.subplots()
ax.plot(women_degrees["Year"],women_degrees["Biology"], label="Women", color="green")
ax.plot(women_degrees["Year"],100-women_degrees["Biology"], label="Men", color="blue")
ax.set_title("Percentage of Biology Degrees Awarded By Gender")
ax.tick_params(left="off",bottom="off",top="off",right="off")
for each in ax.spines:
    ax.spines[each].set_visible(False)
plt.legend(loc="upper right")
```

```python
# multiple plots

cols = ["Biology","Computer Science","Engineering","Math and Statistics"]

fig = plt.figure(figsize=(10,10))
for i, each in enumerate(cols):
    ax = fig.add_subplot(2,2,i+1)
    ax.plot(women_degrees["Year"],women_degrees[each], label="Women", color="blue")
    ax.plot(women_degrees["Year"],100-women_degrees[each], label="Men", color="green")
    ax.tick_params(top="off",bottom="off",left="off",right="off")
    ax.set_ylim(0,100)
    ax.set_xlim(women_degrees.iloc[0]["Year"],women_degrees.iloc[-1]["Year"])
    ax.set_title(each)
    for key in ax.spines:
        ax.spines[key].set_visible(False)

plt.legend(loc="upper right")
```

## Color, Layout, and Annotations

```python
# changing colors to color blind friendly

cols = ["Biology","Computer Science","Engineering","Math and Statistics"]

fig = plt.figure(figsize=(12,12))
dark_blue = "#006BA4"
orange = (255/255,128/255,14/255)
for i, each in enumerate(cols):
    ax = fig.add_subplot(2,2,i+1)
    ax.plot(women_degrees["Year"],women_degrees[each], label="Women", color="#006BA4")
    ax.plot(women_degrees["Year"],100-women_degrees[each], label="Men", color=orange)
    ax.tick_params(top="off",bottom="off",left="off",right="off")
    ax.set_ylim(0,100)
    ax.set_xlim(women_degrees.iloc[0]["Year"],women_degrees.iloc[-1]["Year"])
    ax.set_title(each)
    for key in ax.spines:
        ax.spines[key].set_visible(False)

plt.legend(loc="upper right")
plt.show()
```

```python
# changing line width

cols = ["Biology","Computer Science","Engineering","Math and Statistics"]

fig = plt.figure(figsize=(12,12))
dark_blue = "#006BA4"
orange = (255/255,128/255,14/255)
for i, each in enumerate(cols):
    ax = fig.add_subplot(2,2,i+1)
    ax.plot(women_degrees["Year"],women_degrees[each], label="Women", color="#006BA4", linewidth=3)
    ax.plot(women_degrees["Year"],100-women_degrees[each], label="Men", color=orange, linewidth=3)
    ax.tick_params(top="off",bottom="off",left="off",right="off")
    ax.set_ylim(0,100)
    ax.set_xlim(women_degrees.iloc[0]["Year"],women_degrees.iloc[-1]["Year"])
    ax.set_title(each)
    for key in ax.spines:
        ax.spines[key].set_visible(False)

plt.legend(loc="upper right")
plt.show()
```

```python
# all in a line

stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18,3))
dark_blue = "#006BA4"
orange = (255/255,128/255,14/255)
for i, each in enumerate(stem_cats):
    ax = fig.add_subplot(1,6,i+1)
    ax.plot(women_degrees["Year"],women_degrees[each], label="Women", color="#006BA4", linewidth=3)
    ax.plot(women_degrees["Year"],100-women_degrees[each], label="Men", color=orange, linewidth=3)
    ax.tick_params(top="off",bottom="off",left="off",right="off")
    ax.set_ylim(0,100)
    ax.set_xlim(women_degrees.iloc[0]["Year"],women_degrees.iloc[-1]["Year"])
    ax.set_title(each)
    if i == 0:
        ax.text(2005,87,"Men")
        ax.text(2002,8,"Women")
    if i == len(stem_cats)-1:
        ax.text(2005,62,"Men")
        ax.text(2001,35,"Women")
    for key in ax.spines:
        ax.spines[key].set_visible(False)



plt.show()
```

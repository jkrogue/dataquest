# Bar plots and scatter plots mission

### Using fandango_score_comparison.csv

* FILM - film name
* RT_user_norm - average user rating from Rotten Tomatoes, normalized to a 1 to 5 point scale
* Metacritic_user_nom - average user rating from Metacritc, normalized to a 1 to 5 point scale
* IMDB_norm - average user rating from IMDB, normalized to a 1 to 5 point scale
* Fandango_Ratingvalue - average user rating from Fandango, normalized to a 1 to 5 point scale
* Fandango_Stars - the rating displayed on the Fandango website (rounded to nearest star, 1 to 5 point scale)

```python
# 2/10

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

norm_reviews = pd.read_csv("data/fandango_score_comparison.csv")

cols_to_select = ["FILM","RT_user_norm","Metacritic_user_nom","IMDB_norm","Fandango_Ratingvalue","Fandango_Stars"]

norm_reviews = norm_reviews[cols_to_select]

print(norm_reviews.loc[0])
```

```python
#4/10
num_cols = ["RT_user_norm","Metacritic_user_nom","IMDB_norm","Fandango_Ratingvalue","Fandango_Stars"]

bar_heights = norm_reviews[num_cols].iloc[0].values
fig, ax = plt.subplots()
ax.bar(left = np.arange(5) + 0.75, height = bar_heights, width = 0.5)
```

```python
#5/10
fig,ax = plt.subplots()
bar_heights = norm_reviews[num_cols].iloc[0].values
ax.bar(left = np.arange(5) + 1, height = bar_heights, width = 0.50)
ax.set_xticks(range(1,6))
ax.set_xticklabels(num_cols, rotation=90, label="Rating Source")
ax.set_ylabel("Average Rating")
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015)")
```

```python
# 6/10

fig, ax = plt.subplots()
bar_positions = np.arange(1,6)
bar_widths = norm_reviews[num_cols].iloc[0].values
ax.barh(y=bar_positions,width=bar_widths,height=0.5)
ax.set_yticks(range(1,6))
ax.set_yticklabels(num_cols)
ax.set_ylabel("Rating source")
ax.set_xlabel("Average rating")
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015)")
```

```python
# 7/10 scatterplots

fig, ax = plt.subplots()
ax2 = fig.add_subplot(111)
ax2.scatter(x=norm_reviews["Fandango_Ratingvalue"],y=norm_reviews["RT_user_norm"])
ax2.set_ylabel("Rotten Tomatoes")
ax2.set_xlabel("Fandango")
plt.show()
```

```python
fig = plt.figure(figsize = (5,10))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax1.scatter(x=norm_reviews["Fandango_Ratingvalue"],y=norm_reviews["RT_user_norm"])
ax1.set_ylabel("Rotten Tomatoes")
ax1.set_xlabel("Fandango")

ax2.scatter(x=norm_reviews["RT_user_norm"], y=norm_reviews["Fandango_Ratingvalue"])
ax2.set_ylabel("Fandango")
ax2.set_xlabel("Rotten Tomatoes")

plt.show()
```

```python
fig = plt.figure(figsize=(5,10))

ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)

ax1.scatter(x=norm_reviews["Fandango_Ratingvalue"],y=norm_reviews["RT_user_norm"])
ax1.set_xlabel("Fandango")
ax1.set_ylabel("Rotten Tomatoes")
ax1.set_xlim(0,5)
ax1.set_ylim(0,5)

ax2.scatter(x=norm_reviews["Fandango_Ratingvalue"],y=norm_reviews["Metacritic_user_nom"])
ax2.set_xlabel("Fandango")
ax2.set_ylabel("Metacritic")
ax2.set_xlim(0,5)
ax2.set_ylim(0,5)

ax3.scatter(x=norm_reviews["Fandango_Ratingvalue"],y=norm_reviews["IMDB_norm"])
ax3.set_xlabel("Fandango")
ax3.set_ylabel("IMDB")
ax3.set_xlim(0,5)
ax3.set_ylim(0,5)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

unrate = pd.read_csv("data/UNRATE.csv")
unrate["DATE"] = pd.to_datetime(unrate["DATE"])

plt.plot(unrate["DATE"].iloc[:12],unrate["UNRATE"].iloc[:12])
plt.xticks(rotation=90)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

```python

unrate_1948 = unrate[unrate["DATE"].dt.year == 1948]
unrate_1949 = unrate[unrate["DATE"].dt.year == 1949]
unrate_1950 = unrate[unrate["DATE"].dt.year == 1950]
unrate_1951 = unrate[unrate["DATE"].dt.year == 1951]
unrate_1952 = unrate[unrate["DATE"].dt.year == 1952]

fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(511)
ax2 = fig.add_subplot(512)
ax3 = fig.add_subplot(513)
ax4 = fig.add_subplot(514)
ax5 = fig.add_subplot(515)

ax1.plot(unrate_1948["DATE"], unrate_1948["UNRATE"])
ax2.plot(unrate_1949["DATE"], unrate_1949["UNRATE"])
ax3.plot(unrate_1950["DATE"], unrate_1950["UNRATE"])
ax4.plot(unrate_1951["DATE"], unrate_1951["UNRATE"])
ax5.plot(unrate_1952["DATE"], unrate_1952["UNRATE"])


plt.show()
```

```python
unrate["MONTH"] = unrate["DATE"].dt.month

val_1948 = unrate[unrate["DATE"].dt.year == 1948]
val_1949 = unrate[unrate["DATE"].dt.year == 1949]

fig = plt.figure(figsize = (10,6))
plt.plot(val_1948["MONTH"], val_1948["UNRATE"], c='#FF0000', label=val_1948.iloc[0]["DATE"].year)
plt.plot(val_1949["MONTH"], val_1949["UNRATE"], c=(0,0,1), label="1949")
val_1950 = unrate[unrate["DATE"].dt.year == 1950]
val_1951 = unrate[unrate["DATE"].dt.year == 1951]
val_1952 = unrate[unrate["DATE"].dt.year == 1952]

plt.plot(val_1950["MONTH"], val_1950["UNRATE"], c='#00FF00', label="1950")
plt.plot(val_1951["MONTH"], val_1951["UNRATE"], c="orange", label="1951")
plt.plot(val_1952["MONTH"], val_1952["UNRATE"], c="black", label="1952")
plt.legend(loc="upper left")

plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel("Month, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.show()
```
